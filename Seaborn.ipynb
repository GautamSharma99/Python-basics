1. RELATIONAL PLOT:

  . to see the statistical relation between 2 or more variables
  . bivariate analysis

  PLOTS UNDER THIS:
  
  . scatterplot
  . lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatter plot -> axes level function

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
# relplot -> figure level -> sqaure shape -> 
# almost same as scatter plot

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country']=='India']
temp_df
```

```python
# axis level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# fig level function
# using relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df1 = gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
sns.relplot(kind='line',data=temp_df1,x='year',y='lifeExp',hue='country',style='continent')
```

```python
# facet plot -> plots different graphs acc to the categories in that particular column or row
# fig level function -> doesnot work on axes level

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='smoker')
```

```python
# col wrap -> used to control how many graphs should be in one row

sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

2. DISTRIBUTION PLOTS:

   . used for univariate analysis
   . used to find out the distribution
   . range of the observation
   . central tendency
   . is the data bimodal?
   . are there outliers?

   PLOTS UNDER DISTRIBUTION PLOT:  

   . histplot
   . kdeplot
   . rugplot 

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

```python
sns.displot(data=tips,x='day',kind='hist')
```

```python
sns.displot(data=tips,x='day',kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
# faceting using col and row -> not works on axes level

sns.displot(data=tips,x='tip',kind='hist',col='sex',element='step')
```

```python
# kdeplot(kernel density estimate)
# rather than using discrete bins, a KDE plot smooths the observations with a gaussian kernel,producing a continous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
# fig level
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue and fill

sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# rugplot (represents the density)

# plot marginal distribution by drawing ticks along the x and y axes
# this function is intended to complement other plots by showing the location of individual observations in an unobstructive way

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate kdeplot
# a bivariate histogram bins the data within rectangles that title the plot
# and then shows the count of obseravations within each rectangle with the fill color 

sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
# bivariate kdeplot
# a bivariate KDE plot smoothes the (x,y) observation with a 2D Gaussian

sns.kdeplot(data=tips,x='total_bill',y='tip')
```
