2D line plot:

. Bivaraite analysis
. categorical -> numerical and numerical -> numerical
. use case - Time series data

```python
 # plotting a simple function
price = [48000,54000,57000,47000,45000,60000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
batsman = pd.read_csv("C:/Users/gauta/Downloads/sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# label 
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='Green',linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma and virat Kohli IPL carrier comparison')

```

```python
# legend

plt.plot(batsman['index'],batsman['V Kohli'],color='Green',linestyle='solid',linewidth=3,marker='D',label='virat kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashdot',linewidth=2,marker='o',label='rohit sharma')

plt.title('Rohit Sharma and virat Kohli IPL carrier comparison')

plt.legend()
```

```python
# limiting axes
price = [48000,54000,57000,47000,45000,60000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
# in these cases when one value is very large than others,the points seems straight. to tackle this we use xlim and ylim
plt.ylim(0,75000)
plt.xlim(2016,2019)
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='Green',linestyle='solid',linewidth=3,marker='D',label='virat kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashdot',linewidth=2,marker='o',label='rohit sharma')

plt.title('Rohit Sharma and virat Kohli IPL carrier comparison')

plt.legend()

plt.grid()
```

SUBPLOTS:

```python
# A diff way to plot graphs
batters.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('something')
ax.set_xlabel('avg')
ax.set_ylabel('strike rate')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('avg vs strike rate')
ax[0].set_ylabel('strike rate')

ax[1].set_ylabel('runs')
ax[1].set_title('avg vs runs')
ax[1].set_xlabel('avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'])
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
# another way to do the exact same thing

fig = plt.figure()

ax1 = fig.add_subplot(3,1,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(3,1,2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(3,1,3)
ax3.hist(batters['avg'])

```

3D SURFACE PLOTS:

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)

z = xx**2 + y**2
```

```python

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z)
```
